Using R magic in Jupyter, load the `iris` data set from R into a `pandas` DataFrame and use `seaborn` to generate the plot shown.

![img](./iris.png)

```python
%matplotlib inline

import matplotlib as plt
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris = pd.melt(iris, id_vars = 'species', var_name = 'measurement', value_name = 'value')
iris.head()
```

```python
sns.catplot(x='measurement', y='value', data=iris, col = 'species', kind = 'violin')
pass
```
