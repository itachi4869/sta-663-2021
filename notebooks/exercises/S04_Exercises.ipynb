**Exercise 1**.

- Create the `mtcars` DataFrame from this URL `https://bit.ly/3oAZyrd`
- Find the mean `wt` by number of cylinders sorted in decreasing order, keeping the result as a DataFrame.
- Find all car models that have numbers in the name
- Create a correlation matrix of `hp`, `weight` and `qsec` for cars with 4 gears

```python
url = 'https://bit.ly/3oAZyrd'
```

```python
import pandas as pd

df = pd.read_csv(url)
df.head()
```

```python
df.groupby('cyl')[['wt']].mean().sort_values(by = 'cyl', ascending = False)
```

```python
df[df.model.str.contains(r'[0-9]')][['model']]
```

```python
df[df.gear == 4][['hp', 'wt', 'qsec']].corr()
```

**Exercise 2**.

- Convert the data in `mess.txt` to the DataFrame in `tidy.csv` using Pandas operations


```python
! cat mess.txt
```

```python
df = pd.read_csv('mess.txt', sep = ' ')
```

```python
df[['name','age', 'site']] = df.desc.str.split('-', expand = True)
df = df.drop(columns = 'desc')
df
```

```python
df = pd.melt(df, id_vars=['name', 'age', 'site'], var_name='gene', value_name='expression')
df['gene'] = df.gene.str.replace('gene', '')
df.sort_values('name')
```
